```python
import numpy as np
import matplotlib.pyplot as plt

def fade(t):
    """Smoothstep function for interpolation."""
    return t * t * t * (t * (t * 6 - 15) + 10)

def lerp(a, b, t):
    """Linear interpolation."""
    return a + t * (b - a)

def perlin1d(x, seed=0):
    """Simple 1D Perlin noise function."""
    np.random.seed(seed)
    
    # Create a random gradient at integer points
    grid_x = np.floor(x).astype(int)
    grad = np.random.uniform(-1, 1, len(grid_x) + 1)  # Random gradients
    
    # Compute fade interpolation value
    t = fade(x - grid_x)
    
    # Perform linear interpolation between two gradient values
    return lerp(grad[grid_x], grad[grid_x + 1], t)

# Generate 1D Perlin noise
x = np.linspace(0, 10, 500)
y = perlin1d(x)

# Plot the Perlin noise
plt.figure(figsize=(10, 4))
plt.plot(x, y, label="Perlin Noise")
plt.legend()
plt.title("1D Perlin Noise")
plt.show()
```
